```python
# 從coliee_dataset/task1/task1_train_labels_2025.json中讀入資料
import json
import os
# 如果檔案不存在，就報錯停止
file_path = "../../coliee_dataset/task1/task1_train_labels_2025.json"
if not os.path.exists(file_path):
    raise FileNotFoundError(f"{file_path} 不存在")

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)
    # 檢查讀入json data的頂層結構
    if type(data) is dict:
        print("頂層結構是 {}")
    elif type(data) is list:
        print("頂層結構是 []")
    else:
        print("頂層結構不是 {} 或 []")
print(f"成功讀取 {file_path}，共有 {len(data)} 筆資料")
```

```python
# 顯示前3筆data的內容
for k,v in list(data.items())[:3]:
    print(f"key: {k}")
    print(f"value: {v}")
    print("-----")
```

```python
# 隨機從key中挑選1個
import random
random_key = random.choice(list(data.keys()))
print(f"隨機挑選的key: {random_key}")
print(f"對應的value: {data[random_key]}")
```
